# データの永続化
## SQLite
* 軽量なデータベースを実袋するc言語のライブラリ
* サーバようのプロセスを必要としない、組み込み型のデータベースエンジン。
* 単ーのファイルによって管理される（インメモリも可能）

```python
import sqlite3
```

```python
#DB ファイルの保存先パス (相対パスで指定)#DB file save path (specify relative path)
path = ''
# googleColab の場合
# path = '/content/'

# DB ファイル
db_name = 'test.db'
# DB 接続の確立　Establishing a DB connection
# DB ファイルが存在しない場合は、新規作成される　// if the DB file dsoes not exist , it will be creataed
conn = sqlite3.connect(path + db_name)
# DB への接続を閉じる // closing the connection to the DB
conn.close()
```

# テーブルの作成

```python
path = ''
db_name = 'car.db'
try:
    
    # DB 接続オブジェクトの作成 // creating a Db connection object
    conn = sqlite3.connect(path + db_name)

    # SQL (RDB　を　操作するための言語)　を　実行するためのカーソルオブジェクトを取得 // creating a cursor object to execute the SQL(language to operate RDB)
    cur = conn.cursor()

    # SQL 文の作成 // creating an SQL statement
    # テーブルの作成　Creating a table
    sql = 'CREATE TABLE cars (id INT, name TEXT, price REAL);'
    # SQL 分の実行 executing the SQL statement
    cur.execute(sql)
# DB への接続を閉じる　cloosing the connection to the DB
except sqlite3.Error as e:
    print(f" エラーが発生しました：{e}")
finally:
    conn.close()
    



```

```python
path = ''
db_name = 'car.db'
try:

    # DB 接続オブジェクトの作成 // creating a Db connection object
    conn = sqlite3.connect(path + db_name)

    # SQL (RDB　を　操作するための言語)　を　実行するためのカーソルオブジェクトを取得 // creating a cursor object to execute the SQL(language to operate RDB)
    cur = conn.cursor()

    # SQL 文の作成 // creating an SQL statement
    #   データの輸入　INseting data
    sql = "INSERT INTO cars (id, name, price) VALUES (?, ?, ?);"
    cars = [
        (2, 'bmw', 30000),
        (3, 'benz', 40000),
        (4, 'toyota', 25000 )
    ]
    # テーブルの作成　Creating a table
    # SQL 分の実行 executing the SQL statement
    cur.executemany(sql,cars)
    
    # DB への接続を閉じる　cloosing the connection to the DB

    conn.commit()   # 変更を保存
except sqlite3.Error as e:
    print(f" エラーが発生しました：{e}")
finally:
    conn.close()
```

# # DB内のデータを参照する

```python
# DB内のデータを参照する
path = ''
db_name = 'car.db'
try:

    # DB 接続オブジェクトの作成 // creating a Db connection object
    conn = sqlite3.connect(path + db_name)

    # SQL (RDB　を　操作するための言語)　を　実行するためのカーソルオブジェクトを取得 // creating a cursor object to execute the SQL(language to operate RDB)
    cur = conn.cursor()

    # データを参照するSQL
    sql = "SELECT * FROM cars;"
# SQL 分の実行 executing the SQL statement
    cur.execute(sql)
    
    # DB への接続を閉じる　cloosing the connection to the DB
    # conn.commit()   # 変更を保存
except sqlite3.Error as e:
    print(f" エラーが発生しました：{e}")
else:
    for row in cur:
        id, name, price = row
        print(f"{id},{name},{price} ")
finally:
    conn.close()
```
